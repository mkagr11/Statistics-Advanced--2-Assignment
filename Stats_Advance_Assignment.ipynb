# 1.What is hypothesis testing in statistics
- Hypothsis tetsing a statistical method to determine wheather there is enough evidence to reject the Null hypothesis againts Alternate hypothisis, against a pre-defined significance level.

# 2. What is the null hypothesis, and how does it differ from the alternative hypothesis
- Null hypothsis assumes no difference,no affect.(e.g. TV ad has no impact on Sales)
- Alternate hypothesis suggests a real effect or difference exists (e.g.TV ad has significant impact on sales)

# 3.What is the significance level in hypothesis testing, and why is it important
- Significance level in hypothesis testing is the probability of rejecting the null hypothesis. It is important because it controls the Type I error.

# 4. What does a P-value represent in hypothesis testing
-The P-value in hypothesis testing represents the probability of obtaining results.If p-value < Level of significance, then reject Null hypothesis; otherwise, fail to reject Null hypothesis.

# 5. How do you interpret the P-value in hypothesis testing
-A small P-value (≤ 0.05) suggests strong evidence against the null hypothesis, while a large P-value (> 0.05) indicates weak evidence, meaning we fail to reject the null hypothesis

# 6. What are Type 1 and Type 2 errors in hypothesis testing
-A Type 1 error (False Positive) occurs when we reject a true null hypothesis,
 Type 2 error (False Negative) happens when we fail to reject a false null hypothesis.

# 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing
-A one-tailed test checks for an effect in only one direction of the mean. A two-tailed test checks for an effect in both directions,mean is either significantly higher or lower.

# 8.What is the Z-test, and when is it used in hypothesis testing
- Z -test is used to dtermine wheather sample mean is significatly different from population mean, data must follow nirmal distribution, and sample size is more than 30. It is used when population standard deviation is known.

# 9.How do you calculate the Z-score, and what does it represent in hypothesis testing
- Z-score= Sample mean(X)-μ(population mean)/σ (standard deviation)

# 10.What is the T-distribution, and when should it be used instead of the normal distribution
- when Standard deviation of the population is unkown,and sample size is less than 30.

# 11.What is the difference between a Z-test and a T-test
- A Z-test is used for large samples (n ≥ 30) with known population variance, while a T-test is used for small samples (n < 30) with unknown population variance.

# 12.What is the T-test, and how is it used in hypothesis testing
-A T-test is a statistical test used in hypothesis testing to compare the means of one or two groups, especially when the sample size is small (n < 30) and the population variance is unknown.

# 13.What is the relationship between Z-test and T-test in hypothesis testing
-Both Z-test and T-test compare means in hypothesis testing, but the Z-test is used for large samples (n ≥ 30) with known variance, while the T-test is used for small samples (n < 30) with unknown variance

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
# interpret the results

# --------------Z test Considering fictitios data--------------------------------

# '''## Problem 1: Sales Performance of a New Strategy
# A retail company implemented a new sales strategy and wants to check if it significantly increased the average monthly revenue.
# - Null Hypothesis (\(H_0\)): The average monthly revenue before and after implementing the strategy is the same (\(\mu = 100,000\)).
# - Alternative Hypothesis (\(H_1\)): The new strategy **increased** the average monthly revenue (\(\mu > 100,000\)).
# - Sample mean revenue after implementation: $102,500$
# - Population standard deviation: $4,500$
# - Sample size: **36 months**
# - Significance level (\(\alpha\)): 5% (0.05)
# Find the Z-score, p-value, and conclude whether the strategy improved sales.'''
import numpy as np
from scipy import stats
miu=100000
x=102500
stnad_dev=4500
n=36
alpha=0.05
z_score=(x-miu)/(stnad_dev/np.sqrt(n))
z_critical=stats.norm.ppf(1-alpha)
z_score_probability=1-(stats.norm.cdf(z_score))
print(f"z_score is: {z_score}")
print(f"z_critical is: {z_critical}")
print(f"z_probability is: {z_score_probability}")

#Z critical approach
if z_score_probability<alpha:
    print("Reject the null hypothesis : The new strategy has increased the average monthly revenue")
else:
    print("Fail to reject the null hypothesis: The average monthly revenue is same")
# P value approach
if z_score_probability<alpha:
    print("Reject the null hypothesis : The new strategy has increased the average monthly revenue")
else:
    print("Fail to reject the null hypothesis: The average monthly revenue is same")
# Compare p-value with alpha:
# If p<alpha , then reject H0 (significant change)
# if p>alpha, then fail to reject H0 (no significant change)
```

```python
# 5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_type_errors(mu0, mu1, sigma, n, alpha):
    se = sigma / np.sqrt(n)
    z_crit = stats.norm.ppf(1 - alpha/2)
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    pdf_H0 = stats.norm.pdf(x, mu0, se)
    pdf_H1 = stats.norm.pdf(x, mu1, se)

    plt.figure(figsize=(8,5))
    plt.plot(x, pdf_H0, label='H0', color='blue')
    plt.plot(x, pdf_H1, label='H1', color='orange')

    x_reject_left = x[x <= mu0 - z_crit * se]
    x_reject_right = x[x >= mu0 + z_crit * se]
    plt.fill_between(x_reject_left, stats.norm.pdf(x_reject_left, mu0, se), color='blue', alpha=0.3)
    plt.fill_between(x_reject_right, stats.norm.pdf(x_reject_right, mu0, se), color='blue', alpha=0.3)

    x_accept = x[(x > mu0 - z_crit * se) & (x < mu0 + z_crit * se)]
    plt.fill_between(x_accept, stats.norm.pdf(x_accept, mu1, se), color='orange', alpha=0.3)

    beta = stats.norm.cdf(mu0 + z_crit * se, loc=mu1, scale=se) - stats.norm.cdf(mu0 - z_crit * se, loc=mu1, scale=se)

    plt.legend()
    plt.title(f"Type I Error (alpha): {alpha}, Type II Error (beta): {beta:.3f}")
    plt.xlabel("Test Statistic")
    plt.ylabel("Density")
    plt.show()


plot_type_errors(mu0=0, mu1=0.5, sigma=1, n=30, alpha=0.05)
```

```python
# 7.Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


np.random.seed(0)
n = 30
pre = np.random.normal(loc=100, scale=10, size=n)
post = pre + np.random.normal(loc=5, scale=5, size=n)


t_stat, p_val = stats.ttest_rel(pre, post)
print("t-statistic:", t_stat)
print("p-value:", p_val)


plt.figure(figsize=(8,6))
plt.scatter(np.zeros(n), pre, color='blue', label='Pre')
plt.scatter(np.ones(n), post, color='orange', label='Post')


for i in range(n):
    plt.plot([0, 1], [pre[i], post[i]], color='gray', linestyle='--', alpha=0.5)

plt.xticks([0, 1], ['Pre', 'Post'])
plt.ylabel('Measurement')
plt.title(f'Paired Sample Comparison\n t-statistic: {t_stat:.2f}, p-value: {p_val:.3f}')
plt.legend()
plt.show()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    sem = stats.sem(sample)  # Standard error of the mean
    margin = sem * stats.t.ppf((1 + confidence) / 2.0, n - 1)
    return mean - margin, mean + margin

data = np.random.normal(loc=50, scale=5, size=30)
ci = confidence_interval(data, confidence=0.95)
print("95% Confidence Interval:", ci)
```

```python
# 16. Implement an F-test using Python to compare the variances of two random samples
import numpy as np
import scipy.stats as stats

np.random.seed(0)
sample1 = np.random.normal(loc=0, scale=2, size=30)
sample2 = np.random.normal(loc=0, scale=3, size=30)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)
df1 = len(sample1) - 1
df2 = len(sample2) - 1

if var1 > var2:
    F = var1 / var2
    dfn, dfd = df1, df2
else:
    F = var2 / var1
    dfn, dfd = df2, df1

p_value = 2 * min(stats.f.cdf(F, dfn, dfd),1 - stats.f.cdf(F, dfn, dfd))

print("F-statistic:", F)
print("p-value:", p_value)
```

```python
# 17. Write a Python program to perform an ANOVA test to compare means between multiple groups and
# interpret the results
import numpy as np
import scipy.stats as stats

np.random.seed(0)
group1 = np.random.normal(10, 2, 30)
group2 = np.random.normal(12, 2, 30)
group3 = np.random.normal(11, 2, 30)

F_stat, p_val = stats.f_oneway(group1, group2, group3)
print("ANOVA F-statistic:", F_stat)
print("p-value:", p_val)

if p_val < 0.05:
    print("Reject null hypothesis: Significant difference among group means.")
else:
    print("Fail to reject null hypothesis: No significant difference among group means.")
```

```python
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
import scipy.stats as stats

def check_anova_assumptions(groups):
    normality_results = [stats.shapiro(g) for g in groups]
    equal_variance_result = stats.levene(*groups)
    return normality_results, equal_variance_result

group1 = np.random.normal(10, 2, 30)
group2 = np.random.normal(12, 2, 30)
group3 = np.random.normal(11, 2, 30)

normality, equal_var = check_anova_assumptions([group1, group2, group3])
print("Normality test results (W, p):", normality)
print("Equal variance test result (W, p):", equal_var)
```

```python
# 20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
# results
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
from statsmodels.graphics.factorplots import interaction_plot

np.random.seed(42)
n = 15
factorA = np.repeat(['A1', 'A2'], n*2)
factorB = np.tile(np.repeat(['B1', 'B2'], n), 2)
response = np.random.randn(len(factorA)) + np.where(factorA=='A1', 0, 1) + np.where(factorB=='B1', 0, 1)
df = pd.DataFrame({'factorA': factorA, 'factorB': factorB, 'response': response})

model = smf.ols('response ~ C(factorA) * C(factorB)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)

plt.figure(figsize=(8,6))
interaction_plot(df['factorA'], df['factorB'], df['response'], colors=['blue','red'])
plt.xlabel('Factor A')
plt.ylabel('Response')
plt.title('Interaction Plot')
plt.show()
```

```python
# 24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats

np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=20)
sigma0_sq = 25
n = len(data)
s_sq = np.var(data, ddof=1)
chi2_stat = (n - 1) * s_sq / sigma0_sq
df = n - 1
lower_tail = stats.chi2.cdf(chi2_stat, df)
upper_tail = 1 - stats.chi2.cdf(chi2_stat, df)
p_value = 2 * min(lower_tail, upper_tail)

print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("Reject the null hypothesis: Population variance differs.")
else:
    print("Fail to reject the null hypothesis: Insufficient evidence of variance difference.")
```

```python
# 25.Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import scipy.stats as stats
success1, n1 = 45, 100
success2, n2 = 30, 100

p1 = success1 / n1
p2 = success2 / n2
p_pool = (success1 + success2) / (n1 + n2)
se = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))
z_stat = (p1 - p2) / se
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Z-statistic:", z_stat)
print("p-value:", p_value)
```
